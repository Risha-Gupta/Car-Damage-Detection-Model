```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV data
df = pd.read_csv("traindata.csv")

# Write to Excel
df.to_excel("train_metrics.xlsx", index=False)

# Plot Loss Curves
plt.figure(figsize=(10,6))
plt.plot(df['Epoch'], df['Box_Loss'], label='Box Loss')
plt.plot(df['Epoch'], df['Seg_Loss'], label='Seg Loss')
plt.plot(df['Epoch'], df['Cls_Loss'], label='Cls Loss')
plt.plot(df['Epoch'], df['Dfl_Loss'], label='Dfl Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss Value')
plt.title('Loss Curves')
plt.legend()
plt.grid()
plt.savefig('loss_curves.png')
plt.close()

# Plot Box mAP50 and mAP50-95
plt.figure(figsize=(10,6))
plt.plot(df['Epoch'], df['Box_mAP50'], label='Box mAP@50')
plt.plot(df['Epoch'], df['Box_mAP50-95'], label='Box mAP@50-95')
plt.xlabel('Epoch')
plt.ylabel('mAP')
plt.title('Box mAP Metrics')
plt.legend()
plt.grid()
plt.savefig('box_map.png')
plt.close()

# Plot Mask mAP50 and mAP50-95
plt.figure(figsize=(10,6))
plt.plot(df['Epoch'], df['Mask_mAP50'], label='Mask mAP@50')
plt.plot(df['Epoch'], df['Mask_mAP50-95'], label='Mask mAP@50-95')
plt.xlabel('Epoch')
plt.ylabel('mAP')
plt.title('Mask mAP Metrics')
plt.legend()
plt.grid()
plt.savefig('mask_map.png')
plt.close()

# Plot Box Precision and Recall
plt.figure(figsize=(10,6))
plt.plot(df['Epoch'], df['Box_P'], label='Box Precision')
plt.plot(df['Epoch'], df['Box_R'], label='Box Recall')
plt.xlabel('Epoch')
plt.ylabel('Score')
plt.title('Box Precision & Recall')
plt.legend()
plt.grid()
plt.savefig('box_precision_recall.png')
plt.close()

print("Excel file and graphs generated successfully.")
```
