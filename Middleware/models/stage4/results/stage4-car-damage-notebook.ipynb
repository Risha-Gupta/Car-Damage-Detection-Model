```python
import os, json
from ultralytics import YOLO
import glob

model_path = '/kaggle/input/stage3/pytorch/default/1/best2.pt'  # or your selected weights
model = YOLO(model_path)
test_dir = '/kaggle/input/car-dd/CarDD_release/CarDD_COCO/test2017'  # adjust if needed
images = sorted(glob.glob(f"{test_dir}/*.jpg"))  # adjust extension if .jpg

results_json = []
for img in images:
    image_name = os.path.basename(img)
    results = model(img)
    for r in results:
        for box in r.boxes:
            x1, y1, x2, y2 = box.xyxy[0].cpu().numpy()
            w, h = x2-x1, y2-y1
            bbox = [float(x1), float(y1), float(w), float(h)]
            cls = int(box.cls[0].cpu().numpy())
            conf = float(box.conf[0].cpu().numpy())
            damage_type = model.names[cls] if hasattr(model, 'names') else str(cls)
            results_json.append({
                "image_name": image_name,
                "bbox": bbox,
                "damage_type": damage_type,
                "confidence": conf
            })

with open('/kaggle/working/stage3_results.json', 'w') as f:
    json.dump(results_json, f, indent=2)
print("Stage 3 results saved!")
```

```python
import os
import json
from torch.utils.data import Dataset, DataLoader
from PIL import Image
import numpy as np
import torch
from augmentation import get_train_aug, get_val_aug

# Config paths for train/val
train_ann = '/kaggle/input/car-dd/CarDD_release/CarDD_COCO/annotations/instances_train2017.json'
val_ann   = '/kaggle/input/car-dd/CarDD_release/CarDD_COCO/annotations/instances_val2017.json'
train_img_dir = '/kaggle/input/car-dd/CarDD_release/CarDD_COCO/train2017'
val_img_dir   = '/kaggle/input/car-dd/CarDD_release/CarDD_COCO/val2017'
img_size = 224

def load_coco_ann(ann_path):
    with open(ann_path) as f:
        data = json.load(f)
    id2name = {cat['id']: cat['name'] for cat in data['categories']}
    imgid2file = {img['id']: img['file_name'] for img in data['images']}
    imgid2label = {}
    for ann in data['annotations']:
        img_id = ann['image_id']
        cat_id = ann['category_id']
        imgid2label[img_id] = id2name[cat_id]
    images = []
    labels = []
    for img_id, file_name in imgid2file.items():
        images.append(file_name)
        labels.append(imgid2label.get(img_id, 'unknown'))
    return images, labels

train_images, train_labels = load_coco_ann(train_ann)
val_images,   val_labels   = load_coco_ann(val_ann)

# --- Albumentations Dataset ---
class SimpleCocoDataset(Dataset):
    def __init__(self, img_dir, files, labels, transform=None):
        self.img_dir = img_dir
        self.files = files
        self.labels = labels
        self.transform = transform
    def __len__(self):
        return len(self.files)
    def __getitem__(self, idx):
        img_path = os.path.join(self.img_dir, self.files[idx])
        image = np.array(Image.open(img_path).convert('RGB'))
        label = self.labels[idx]
        if self.transform:
            transformed = self.transform(image=image)
            image = transformed['image']
        return image, label

train_aug = get_train_aug(img_size)
val_aug = get_val_aug(img_size)

train_dataset = SimpleCocoDataset(train_img_dir, train_images, train_labels, transform=train_aug)
val_dataset   = SimpleCocoDataset(val_img_dir,   val_images,   val_labels,   transform=val_aug)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True, num_workers=2)
val_loader   = DataLoader(val_dataset,   batch_size=32, num_workers=2)

# --- Check sample batch ---
batch = next(iter(train_loader))
print('Batch loaded:', batch[0].shape, batch[1][:5])
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from model import ComponentClassifier
import os
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.metrics import precision_score, recall_score, f1_score

CLASS_NAMES = ['dent', 'scratch', 'crack', 'glass shatter', 'lamp broken', 'tire flat']
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
output_path = '/kaggle/working'
os.makedirs(output_path, exist_ok=True)

model = ComponentClassifier(backbone='efficientnet_b3', num_classes=6, pretrained=True)
model.to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=1e-4, weight_decay=1e-4)

num_epochs = 50
best_val_loss = float('inf')

metrics = {
    'Epoch': [],
    'Box_P': [],
    'Box_R': [],
    'Box_mAP50': [],
    'Box_mAP50-95': [],
    'Mask_P': [],
    'Mask_R': [],
    'Mask_mAP50': [],
    'Mask_mAP50-95': [],
    'Box_Loss': [],
    'Seg_Loss': [],
    'Cls_Loss': [],
    'Dfl_Loss': []
}

def train_one_epoch(model, dataloader, criterion, optimizer, device, class_names):
    model.train()
    running_loss = 0.0
    all_preds = []
    all_labels = []
    for images, labels in dataloader:
        images = images.to(device)
        labels_idx = torch.tensor([class_names.index(l) for l in labels]).to(device)
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels_idx)
        loss.backward()
        optimizer.step()
        running_loss += loss.item() * images.size(0)
        _, preds = torch.max(outputs, 1)
        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels_idx.cpu().numpy())
    avg_loss = running_loss / len(dataloader.dataset)
    precision = precision_score(all_labels, all_preds, average='weighted', zero_division=0)
    recall = recall_score(all_labels, all_preds, average='weighted', zero_division=0)
    accuracy = np.mean(np.array(all_preds) == np.array(all_labels))
    return avg_loss, precision, recall, accuracy

def validate(model, dataloader, criterion, device, class_names):
    model.eval()
    running_loss = 0.0
    all_preds = []
    all_labels = []
    with torch.no_grad():
        for images, labels in dataloader:
            images = images.to(device)
            labels_idx = torch.tensor([class_names.index(l) for l in labels]).to(device)
            outputs = model(images)
            loss = criterion(outputs, labels_idx)
            running_loss += loss.item() * images.size(0)
            _, preds = torch.max(outputs, 1)
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels_idx.cpu().numpy())
    avg_loss = running_loss / len(dataloader.dataset)
    precision = precision_score(all_labels, all_preds, average='weighted', zero_division=0)
    recall = recall_score(all_labels, all_preds, average='weighted', zero_division=0)
    accuracy = np.mean(np.array(all_preds) == np.array(all_labels))
    return avg_loss, precision, recall, accuracy

for epoch in range(num_epochs):
    train_loss, train_p, train_r, train_acc = train_one_epoch(model, train_loader, criterion, optimizer, device, CLASS_NAMES)
    val_loss, val_p, val_r, val_acc = validate(model, val_loader, criterion, device, CLASS_NAMES)
    # Set box/mask metrics and extra losses to zero for classifier
    metrics['Epoch'].append(epoch + 1)
    metrics['Box_P'].append(val_p)
    metrics['Box_R'].append(val_r)
    metrics['Box_mAP50'].append(0)          # Not available in classifier, so filled with 0
    metrics['Box_mAP50-95'].append(0)
    metrics['Mask_P'].append(0)
    metrics['Mask_R'].append(0)
    metrics['Mask_mAP50'].append(0)
    metrics['Mask_mAP50-95'].append(0)
    metrics['Box_Loss'].append(0)
    metrics['Seg_Loss'].append(0)
    metrics['Cls_Loss'].append(val_loss)
    metrics['Dfl_Loss'].append(0)
    print(f"Epoch {epoch+1}/{num_epochs}: Train Loss={train_loss:.4f}, Val Loss={val_loss:.4f}, Val Precision={val_p:.4f}, Val Recall={val_r:.4f}, Val Acc={val_acc:.4f}")
    if val_loss < best_val_loss:
        best_val_loss = val_loss
        torch.save(model.state_dict(), os.path.join(output_path, 'best_model.pt'))
        print(" Saved best model")

df = pd.DataFrame(metrics)
excel_path = os.path.join(output_path, 'stage4_metrics.xlsx')
df.to_excel(excel_path, index=False)
print(f" Metrics saved to {excel_path}")

# --- Graphs identical to your uploads ---
plt.figure(figsize=(10,6))
plt.plot(df['Epoch'], df['Box_mAP50'], label='Box mAP@50')
plt.plot(df['Epoch'], df['Box_mAP50-95'], label='Box mAP@50-95')
plt.xlabel('Epoch')
plt.ylabel('mAP')
plt.title('Box mAP Metrics')
plt.legend()
plt.grid(True)
plt.savefig(os.path.join(output_path, 'box_map.png'))
plt.close()

plt.figure(figsize=(10,6))
plt.plot(df['Epoch'], df['Box_P'], label='Box Precision')
plt.plot(df['Epoch'], df['Box_R'], label='Box Recall')
plt.xlabel('Epoch')
plt.ylabel('Score')
plt.title('Box Precision & Recall')
plt.legend()
plt.grid(True)
plt.savefig(os.path.join(output_path, 'box_precision_recall.png'))
plt.close()

plt.figure(figsize=(10,6))
plt.plot(df['Epoch'], df['Mask_mAP50'], label='Mask mAP@50')
plt.plot(df['Epoch'], df['Mask_mAP50-95'], label='Mask mAP@50-95')
plt.xlabel('Epoch')
plt.ylabel('mAP')
plt.title('Mask mAP Metrics')
plt.legend()
plt.grid(True)
plt.savefig(os.path.join(output_path, 'mask_map.png'))
plt.close()

plt.figure(figsize=(10,6))
plt.plot(df['Epoch'], df['Box_Loss'], label='Box Loss')
plt.plot(df['Epoch'], df['Seg_Loss'], label='Seg Loss')
plt.plot(df['Epoch'], df['Cls_Loss'], label='Cls Loss')
plt.plot(df['Epoch'], df['Dfl_Loss'], label='Dfl Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss Value')
plt.title('Loss Curves')
plt.legend()
plt.grid(True)
plt.savefig(os.path.join(output_path, 'loss_curves.png'))
plt.close()

print("\n ALL METRICS AND GRAPHS GENERATED IN kaggle/working!")
```
